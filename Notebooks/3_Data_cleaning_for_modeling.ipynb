```python
#
```

```python
from collections import Counter
remove_df["PROMISE_STATUS_O"].replace(np.nan,Counter(remove_df["PROMISE_STATUS_O"]).most_common()[0][0],inplace=True)
remove_df["PROMISE_STATUS_O"].isnull().sum()
```

```python
remove_df["EXTENSION_ELIGIBLE_Y"].replace(np.nan,Counter(remove_df["EXTENSION_ELIGIBLE_Y"]).most_common()[0][0],inplace=True)
remove_df["lastchannel_Z"].replace(np.nan,Counter(remove_df["lastchannel_Z"]).most_common()[0][0],inplace=True)
remove_df["REM_TM_RAT_OBS"].replace(np.nan,Counter(remove_df["REM_TM_RAT_OBS"]).most_common()[0][0],inplace=True)

remove_df
```

```python
#Imputed with median
for col in remove_df.columns:
    remove_df[col].replace(np.nan,remove_df[col].median(),inplace=True)
remove_df.isnull().sum()
```

```python
#Outliers
```

```python
q1=remove_df.quantile(0.25)
q3=remove_df.quantile(0.75)
IQR = q3-q1
IQR
```

```python
1.5* IQR["LTVEffective"]+remove_df["LTVEffective"].quantile(0.75)
```

```python
q3=remove_df["LTVEffective"].quantile(0.75)
remove_df.loc[(remove_df["LTVEffective"] > q3+1.5*IQR["LTVEffective"]),"LTVEffective"]= q3+1.5*IQR["LTVEffective"]
remove_df[remove_df["LTVEffective"] > q3+1.5*IQR["LTVEffective"]]["LTVEffective"]
```

```python
for col in remove_df.drop(columns=["PROMISE_STATUS_O","EXTENSION_ELIGIBLE_Y","lastchannel_Z","REM_TM_RAT_OBS","DPD_bool"]).columns:
    remove_df.loc[remove_df[col] > remove_df[col].quantile(0.75)+1.5*IQR[col],col]=remove_df[col].quantile(0.75)+1.5*IQR[col]
    remove_df.loc[remove_df[col] < remove_df[col].quantile(0.25)-1.5*IQR[col],col]=remove_df[col].quantile(0.25)-1.5*IQR[col]
remove_df
```
