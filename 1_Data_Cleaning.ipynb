```python
utd_latest_df_1 = utd_df_1.sort_values(by="PullDate",ascending=False).drop_duplicates("masked_account")
utd_latest_df_1
```

```python
utd_latest_dpd = utd_dpd.sort_values(by="Business_Date").drop_duplicates("Masked_Account")
utd_latest_dpd
```

```python
latest_merge_df = pd.merge(utd_latest_df_1,utd_latest_dpd,left_on="masked_account",right_on="Masked_Account",how="outer")
latest_merge_df
```

```python
latest_merge_df.info()
```

```python
latest_merge_df["ApplicantFICOScore_bins"]=pd.cut(latest_merge_df["ApplicantFICOScore"],[0,300,600,700,800,850,np.inf])
latest_merge_df["ApplicantFICOScore_bins"]
```

```python
latest_merge_df["DPD_bins"]= pd.cut(latest_merge_df["DPD"],[-np.inf,0,15,30,60,90,np.inf])
latest_merge_df["DPD_bins"]
```

```python
latest_merge_df[["ApplicantFICOScore_bins","DPD_bins"]].groupby(["ApplicantFICOScore_bins","DPD_bins"]).size().unstack(fill_value=0)
```
